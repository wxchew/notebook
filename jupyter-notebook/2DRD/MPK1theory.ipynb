```python
%matplotlib inline
import numpy as np
import math
import sys
from numpy import sqrt
from math import pi
import matplotlib.pyplot as plt
import mpmath as mp
from scipy import special
from scipy import integrate
from mpmath import fp
import mpmath as mp
from mpmath import invertlaplace as ilp
mydehoog = mp.calculus.inverselaplace.deHoog(mp)

def G(s):
    taud=R*R/D
    x=mp.sqrt(s*taud)
    return mp.besselk(0,x)/(2*pi*D*x*mp.besselk(1,x))

def kirr(s):
    return ka/s/(1+ka*G(s))

def S(k,c,tt,dt):
    intk=integrate.cumtrapz(k,tt,initial=0)
    return np.exp(-c*intk)

def iLP(fp,TT):    #inverse laplace transform of k(z)
    ft=[]
    mp.dps = 5
    for t in TT:
        out = ilp(fp,t)
        ft.append(out)
        #t = mp.convert(t)
        #mydehoog.calc_laplace_parameter(t,alpha=0,tol=1e-9)
        #fpvec = [fp(p) for p in mydehoog.p]
        #out=mydehoog.calc_time_domain_solution(fpvec,t,manual_prec=False)
        #ft.append(out)
    return np.array([float(f) for f in ft])

def Ss(s):
    stexp=[stt*np.exp(-s*TT) for stt,TT in zip(St,T)]
    return np.trapz(stexp,T)

def mpk1(s):
    fgem=1+ka*G(s)
    firr=c0p*ka*Ss(s)/(1-s*Ss(s))
    fmpk1=(fgem*kd/lamb)+(c0*firr*ka/lamb)
    return fmpk1/(s*fmpk1+lamb)

def ilt(start,end,res,N):
    global dt,T,St
    dt=10**(start+res)
    tt1=np.arange(dt,2*10**end,dt)
    tt2=np.logspace(start,end,N)
    kt=iLP(kirr,tt1)
    St=S(kt,c0p,dt)
    St=np.insert(St,0,1)
    T=np.insert(tt1,0,0)
    ST=iLP(mpk1,tt2)
    return tt2,ST

R=0.01
D=1
c0=1000/10.#53.611
ka=1.25
kd=5
lamb=kd+c0*ka
c0p=lamb/ka
S_inf=kd/lamb

start=-5
end=-4
res=-1
N=10
dt=10**(start+res)
T=np.arange(dt,2*10**end+dt,dt)
tt2=np.logspace(start,end,N)
kt=iLP(kirr,T)
St=S(kt,c0p,T,dt)
```
