```python
from math import sqrt as sqrt
from math import pi as pi
from math import exp as exp
from numpy import log as log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
R=0.01
c=1

def k3d(D,kf):
    D=np.array(D)
    kd=4*pi*R*D
    ka=kf*kd
    re=ka*R/(ka+kd)
    phi=4*pi*re**3*c/3
    kon=4*pi*D*re
    return kon*(1+np.sqrt(3*phi))

```

```python
for kf in [0.01,1,100]:
    D=np.logspace(-3,3,30)
    kss=k3d(D,kf)
    plt.loglog(D,1./kss,label=str(kf))
    plt.xlabel('D')
    plt.ylabel('mean')
plt.legend()
```

```python
def k2d(D,kf):
    D=np.array(D)
    kd=2*pi*D
    ka=kf*kd
    phi=pi*R*R*c
    gamma=0.577
    return kd/(0.5*log(2)-gamma-0.5*log(phi)+1./kf)

for kf in [0.01,1,100]:
    D=np.logspace(-3,3,30)
    kss=k2d(D,kf)
    plt.semilogx(D,1./kss,label=str(kf))
    plt.xlabel('D')
    plt.ylabel('mean')
plt.legend()    
```
