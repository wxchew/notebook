```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
rp=0.005
rd=rp*10
R=rp+rd
for c in [0,0.1,1]:
    f=np.logspace(-2,2,50)  
    P=1./(1+1./f)
    t1=np.sqrt(4*math.pi*R**3)
    t2=np.sqrt(c*P**3)
    ratio=1+t1*t2
    perc=abs(ratio-1)*100
    plt.semilogx(f,perc,'-',label='c='+str(c))
plt.ylabel('(kss-kon)/kon, %')
plt.xlabel('kfac')
plt.legend()
```
