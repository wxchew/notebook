```python
import math
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
phi=np.linspace(0,1,50)
phiLSc=0.5
phiBDc=0.32
```

```python
DeLSim = lambda x: 1-(x/phiLSc)
DeLSm = lambda x: 1-(x/1)
DeBDim = lambda x: 1-(x/phiBDc)
DeLSm2 = lambda x: 1-(x*(DeLSim(x)/DeBDim(x))/1)
```

```python
plt.plot(phi,DeLSim(phi),label='LS immobile')
plt.plot(phi,DeBDim(phi),label='BD immobile')
plt.plot(phi,DeLSm(phi),label='LS mobile')
#plt.plot(phi,DeLSm2(phi),label='LS mobile 2')
plt.legend(loc='best')
plt.ylim(0,1)
```
