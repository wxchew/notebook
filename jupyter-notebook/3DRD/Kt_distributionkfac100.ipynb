```python
def calk(stsim,t,C):
    dst = (stsim[2:]-stsim[:-2])/(t[2:]-t[:-2])
    dst=np.append((stsim[1]-stsim[0])/(t[1]-t[0]),dst)
    dst=np.append(dst,(stsim[-1]-stsim[-2])/(t[-1]-t[-2])  )
    return dst/stsim/-C

def caldkt(stsim,std,C,t):
    dkt=((stsim[2:]+stsim[:-2])*std[1:-1]/stsim[1:-1]+std[2:]+std[:-2])/stsim[1:-1]/(t[2:]-t[:-2])
    dkt=np.append((std[1]+(std[0]*stsim[1])/stsim[0])/stsim[0]/(t[1]-t[0]),dkt)
    dkt=np.append(dkt,(std[-2]+(std[-1]*stsim[-2])/stsim[-1])/(t[-1]-t[-2]))
    return dkt/C

def simplekt(stsim,t,C):
    dt=tt[1:]-tt[:-1]
    return (np.log(stsim[1:])-np.log(stsim[:-1]))/(-(C))/dt

```

```python
from scipy.optimize import curve_fit
import math
D=1
rm=0.005
kd= 4.0*math.pi*D*2*rm
ka = 100*kd
kon = (ka*kd)/(ka+kd)
Bck=ka * (2*rm) / (ka + kd)/math.sqrt(D) / math.sqrt(math.pi)
def stt(t,kon,Bt,N,V):
    return np.exp((kon * (t + Bt * np.sqrt(t)))*(-N/V))
def latkt(t,A,B):
    return A*(1+B/(t)**0.5)  
```

```python
from numpy import sqrt
from scipy import special
fig,ax1=plt.subplots(1,1,figsize=(12,7))

def Phi(xx):
    return special.erfcx(xx)
aeff = ka * (2 * rm) / (ka + kd)
Bck=aeff/math.sqrt(D) / math.sqrt(math.pi)
tau = ((ka*2*rm/(ka+kd))**2)/(D)
def intkt1(tt):
    return [kon*(t+(kd*tau/ka)*(2*ka*sqrt(t/tau/math.pi)/kd-1+Phi(ka*sqrt(t/tau)/kd))) for t in tt]

def intkt2(tt):
    return [kon * (t + 2*Bck * np.sqrt(t)) for t in tt]

def stratioT(tt,dt):
    intkt= [kon * (t + 2*Bck * np.sqrt(t)) for t in tt]
    return [np.exp(-i*N*dt/V) for i in intkt]

def stt(intkt):
    return [np.exp(-i*N/V) for i in intkt]

#plt.errorbar(tt,st,std)
plt.plot(tt,stt(intkt1(tt)),'r-',label='exact')
plt.plot(tt,stt(intkt2(tt)),'c--',label='large time asymp')
#plt.plot(tt,stt(intkt3),'k--',label='large time asymp with additional constant')
plt.legend()
#plt.ylim(0.9,1.1)
plt.ylabel('S(t)')
#plt.savefig("/home/chew/ecellwork/figure/ktexact_approx_withsim",format='png',dpi=300)            
```

```python
def fitst(tt,A,B):
    intk=A*(tt+B*sqrt(tt))
    return [np.exp(-i*N/V) for i in intk]

fig,ax1=plt.subplots(1,1,figsize=(12,7))
popt,pcov = curve_fit(fitst,tt,st,p0=[kon,2*Bck],sigma=std)
plt.errorbar(tt,st,std,color='g')
#plt.plot(tt,st,'.')
plt.loglog(tt,fitst(tt,*popt),'b--',label='fitted')
#plt.plot(tt,stt(intkt1),'r-',label='exact')
plt.legend()
#ax1.set_ylim(0.126,0.134)
er=lambda x,y:(x-y)*100/x
perr = np.sqrt(np.diag(pcov))
r=st-fitst(tt,*popt)
meansq=sum(r**2)
chisq=sum((r/std)**2)
plt.ylim(0.95,1)
plt.xlim(0,0.005)
print('msqer',meansq)
print('kon={:.5},konfit={:.5}pm{:.5},diff={:.5},{:.5}%'.format(kon,popt[0],perr[0],kon-popt[0],er(kon,popt[0])))
print('Btheory={:.5},Bfit={:.5}pm{:.5},diff={:.5},{:.5}%'.format(2*Bck,popt[1],perr[1],2*Bck-popt[1],er(2*Bck,popt[1])))
```
