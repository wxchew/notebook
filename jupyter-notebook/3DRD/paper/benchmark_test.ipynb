```python
from ecell4 import *
%matplotlib inline
import numpy as np
import math
import sys
import time
import matplotlib.pyplot as plt
from ecell4.extra import ensemble
import numpy as np

D = 10
ka = 0.01
kd = 1
kp = 1
NE = 909
NS = 9091
NES = 0
NP = 0
Nt=NE+NS
L = 4.4964
rm = 0.03*0.5
kD = 4*math.pi*2*rm*D
kon = ka*kD/(ka+kD)
voxelr = rm
td = (2*voxelr)**2/(6*D)
print('ka/kD={:.4f}'.format(ka/kD))

def lattice(duration):
    obsdt = duration*0.01    
    rng = GSLRandomNumberGenerator()
    rng.seed(1)
    w = spatiocyte.SpatiocyteWorld(ones()*L,voxelr)
    (c,r,l)=w.shape()
    Nev=(c - 2) * (r - 2) * (l - 2)
    #print('phi',(NE+NS)*100/Nev)
    with species_attributes():
        E |  {'D': str(D), 'radius':str(voxelr)}
        S |  {'D': str(D), 'radius':str(voxelr)}
        ES |  {'D': str(D), 'radius':str(voxelr)}
        P |  {'D': str(D), 'radius':str(voxelr)}

    with reaction_rules():
        E + S == ES | (ka,kd)
        ES > E + P | kp

    m = get_model()
    w = spatiocyte.SpatiocyteWorld(ones()*L,voxelr,rng)
    w.bind_to(m)
    w.add_molecules(Species('E'), NE)
    w.add_molecules(Species('S'), NS)    
    w.add_molecules(Species('ES'), NES)
    w.add_molecules(Species('P'), NP)    
    sim = spatiocyte.SpatiocyteSimulator(w)
    sim.initialize()
    obs=FixedIntervalNumberObserver(obsdt,('E','S','ES','P'))
    sim.run(duration,obs)    
    return obs

def ODE(duration):
    rng = GSLRandomNumberGenerator()
    rng.seed(1)

    with reaction_rules():
        E + S == ES | (kon,kd)
        ES > E + P | kp

    m = get_model()
    w = ode.ODEWorld(ones()*L)
    w.bind_to(m)
    w.add_molecules(Species('E'), NE)
    w.add_molecules(Species('S'), NS)    
    w.add_molecules(Species('ES'), NES)
    w.add_molecules(Species('P'), NP)    
    sim = ode.ODESimulator(w)
    sim.initialize()
    obs=FixedIntervalNumberObserver(obsdt,('E','S','ES','P'))
    sim.run(duration,obs)    
    return obs
```

```python
t1 = time.time()
obslat = lattice(10)
timetaken = time.time()-t1
print('sim time',timetaken,'sec',timetaken/60,'minute')
```

```python
def gfrd(duration):
    obsdt = duration*0.01        
    rng = GSLRandomNumberGenerator()
    rng.seed(1)
    M = max(int(min(cbrt(Nt), L / (2 * rm))), 3)
    with species_attributes():
        E |  {'D': str(D), 'radius':str(rm)}
        S |  {'D': str(D), 'radius':str(rm)}
        ES |  {'D': str(D), 'radius':str(rm)}
        P |  {'D': str(D), 'radius':str(rm)}

    with reaction_rules():
        E + S == ES | (ka,kd)
        ES > E + P | kp
    m = get_model()
    w = egfrd.EGFRDWorld(ones()*L,Integer3(M,M,M))
    w.bind_to(m)
    w.add_molecules(Species('E'), NE)
    w.add_molecules(Species('S'), NS)    
    w.add_molecules(Species('ES'), NES)
    w.add_molecules(Species('P'), NP)    
    sim = egfrd.EGFRDSimulator(w)
    sim.initialize()
    obs=FixedIntervalNumberObserver(obsdt,('E','S','ES','P'))
    sim.run(duration,obs)    
    return obs

t1 = time.time()
obse = gfrd(10)
timetaken = time.time()-t1
print('sim time',timetaken,'sec',timetaken/60,'minute')
```

```python
fig,ax = plt.subplots(figsize=(12,8))
lss = ['-','--']
obss = [obslat]#[obsode]
obss = [obslat,obsode]
sims = ['lattice','mass action']
for i,obs in enumerate(obss):
    ls = lss[i]
    sim = sims[i]
    t,ne,nes,nP,ns=np.array(obs.data()).T
    plt.plot(t,ne,linestyle=ls,label='E_'+sim)
    plt.plot(t,nes,linestyle=ls,label='ES_'+sim)
    plt.plot(t,nP,linestyle=ls,label='P_'+sim)
    plt.plot(t,ns,linestyle=ls,label='S_'+sim)
plt.ylim(0,10000)
plt.legend()
plt.ylabel('Number of molecule')
plt.xlabel('time (s)')
```
