```python
import math
import numpy as np
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
M=[]
trial = 10000
tmol= 1.
maxt = 500*tmol
disp = np.array([[1,0,0],[0,1,0],[0,0,1],[-1,0,0],[0,-1,0],[0,0,-1]])
def rant():
    return np.random.exponential(tmol)
for i in range(trial):
    t=rant()
    x=np.array([0,0,1])      
    while t<maxt:
        ran = np.random.randint(6)#np.random.uniform(0,1,2)
        #if ran[0]>0.5:
        #    d=1
        #else: d= -1
        d = disp[ran]
        #if (x+d)==0:
        if (sum(x+d)== 0):
            M.append(t)
            break
        else: x+=d
        t+= rant()
#uq,uqcnt = np.unique(M,return_counts=True)
#for i,j in zip(uq,uqcnt):
#    print i,float(j)/trial
```

```python
hist, bin_edges = np.histogram(M,bins=np.logspace(-6,3,30))
tt = 0.5*(bin_edges[:-1]+bin_edges[1:])
dt = -(bin_edges[:-1]-bin_edges[1:])
plt.loglog(tt,[float(h)/trial for h in hist]/dt,'.')
z = lambda x: 0.1*(x**-1.5) #t^(-0.5)
plt.plot(np.logspace(-1,3,50),z(np.logspace(-1,3,50)),label='p ~ t^-1.5')
plt.xlabel('step')
plt.ylabel('first return prob density')
```
