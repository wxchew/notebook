```python
from ecell4 import *
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
f1,ax1 = plt.subplots(1)
f2,ax2 = plt.subplots(1)

def plotmsd(phi):
    dat = np.loadtxt('data/2Dcrowding_LS_phi0.2testrun20.csv',delimiter=',')   
    msd = dat[:,1]
    t = dat[:,0]
    ax1.plot(t,msd,label=str(phi))
    D = 1
    ax1.plot(t,4*D*t,'g--')
    Dt = msd[1:]/t[1:]/4/D
    ax2.loglog(t[1:],Dt,label=str(phi))
    deff = Dt[(t>.1)[1::]]
    return np.mean(deff)

de = []
for p in [0.2]:
    mean = plotmsd(p)
    de.append(mean)
    ax1.set_xlabel("time")
    ax1.set_ylabel("MSD")  
    ax2.set_ylim(0.2,1.2)
    ax2.set_yticks(np.arange(0.2,1.2,0.1))
    ax2.set_xlabel("time")
    ax2.set_ylabel("D/Do")    
    ax1.legend(loc='best')
    ax2.legend(loc='best')  
print mean
```

```python
f1.savefig("figure/2D_diffusion_MSD_t.png",format='png',dpi=100)    
f2.savefig("figure/2D_diffusion_Deff_t.png",format='png',dpi=100)
```
