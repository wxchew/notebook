```python
import glob
import numpy as np
import matplotlib.pyplot as plt
L=700
pattern = '/home/kaizu/src/ecell4-develop/wxchew2017/res'+str(L)+'/keffc_100_'+str(L)+'_*'
filenames = glob.glob(pattern)
data = None

for i in range(10000):
    i+=1
    filename='/home/kaizu/src/ecell4-develop/wxchew2017/res'+str(L)+'/keffc_100_'+str(L)+'_'+str(i)
    if data is None:       
        data = np.loadtxt(filename, delimiter=',')
        yy = data.T[1]
    else:
        data = np.loadtxt(filename, delimiter=',')
        yy=np.vstack((yy,data.T[1]))        
St =np.mean(yy,axis=0)/4000
Std = np.std(yy,axis=0)/4000
t=data.T[0]
```

```python
lag=1
st=St[::lag]
tt=t[::lag]
std=Std[::lag]
```

```python
print(st[:3],tt[:3])
plt.errorbar(tt,st,std)
plt.xlabel('t')
plt.ylabel('S(t)')
```

```python
N=4000
if L==700: 
    V= 43.46455449419419
else:
    V=5.486249703951731
dt=tt[1:]-tt[:-1]
dSt=(np.log(st[1:])-np.log(st[:-1]))/(-(N/V))/dt
E=np.sqrt((std[1:]/st[1:])**2+(std[:-1]/st[:-1])**2)/((N/V))/dt
E2=((std[1:]/st[1:])+(std[:-1]/st[:-1]))/((N/V))/dt
#plt.plot(tt[:-1],dSt,'.')
print(st[:5])
print(dSt[:5])
print(tt[:5])
plt.errorbar(tt[:-1],dSt,E,marker='.')
#plt.xscale('log')
#plt.yscale('log')
plt.xlabel('t')
plt.ylabel('k(t)')
```

```python
from scipy.optimize import curve_fit
import math
D=1
rm=0.005
kd= 4.0*math.pi*D*2*rm
ka = 100*kd
kon = (ka*kd)/(ka+kd)
Bck=ka * (2*rm) / (ka + kd)/math.sqrt(D) / math.sqrt(math.pi)
def stt(t,kon,Bt,N,V):
    return np.exp((kon * (t + Bt * np.sqrt(t)))*(-N/V))
def latkt(t,A,B):
    return A*(1+B/(t)**0.5)  
```

```python
from numpy import sqrt
from scipy import special
fig,ax1=plt.subplots(1,1,figsize=(12,7))

def Phi(xx):
    return special.erfcx(xx)
Bck=ka * (2*rm) / (ka + kd)/math.sqrt(D) / math.sqrt(math.pi)
tau = ((ka*2*rm/(ka+kd))**2)/(D)
intkt1 = [kon*(t+(kd*tau/ka)*(2*ka*sqrt(t/tau/math.pi)/kd-1+Phi(ka*sqrt(t/tau)/kd))) for t in tt]
intkt2= [kon * (t + 2*Bck * np.sqrt(t)) for t in tt]

def stt(intkt):
    return [np.exp(-i*N/V) for i in intkt]

#plt.errorbar(tt,st,std)
plt.plot(tt,stt(intkt1),'r-',label='exact')
plt.plot(tt,stt(intkt2),'c--',label='large time asymp')
#plt.plot(tt,stt(intkt3),'k--',label='large time asymp with additional constant')
plt.legend()
#plt.ylim(0.9,1.1)
plt.ylabel('S(t)')
#plt.savefig("/home/chew/ecellwork/figure/ktexact_approx_withsim",format='png',dpi=300)            
```

```python
aeff = ka * (2 * rm) / (ka + kd)
def fitst(tt,A,B):
    intk=A*(tt+B*sqrt(tt))
    return [np.exp(-i*N/V) for i in intk]

fig,ax1=plt.subplots(1,1,figsize=(12,7))
popt,pcov = curve_fit(fitst,tt,st,p0=[kon,2*Bck],sigma=std)
plt.errorbar(tt,st,std,color='g')
#plt.plot(tt,st,'.')
plt.loglog(tt,fitst(tt,*popt),'b--',label='fitted')
#plt.plot(tt,stt(intkt1),'r-',label='exact')
plt.legend()
#ax1.set_ylim(0.126,0.134)
er=lambda x,y:(x-y)*100/x
perr = np.sqrt(np.diag(pcov))
r=st-fitst(tt,*popt)
meansq=sum(r**2)
chisq=sum((r/std)**2)
plt.ylim(0.95,1)
plt.xlim(0,0.005)
print(meansq,chisq)
print('kon={:.5},konfit={:.5}pm{:.5},diff={:.5},{:.5}%'.format(kon,popt[0],perr[0],kon-popt[0],er(kon,popt[0])))
print('Btheory={:.5},Bfit={:.5}pm{:.5},diff={:.5},{:.5}%'.format(2*Bck,popt[1],perr[1],2*Bck-popt[1],er(2*Bck,popt[1])))
#print('Ctheory={:.5},Cfit={:.5}pm{:.5},diff={:.5}%'.format(-aeff**2 *kd/D/ka,popt[2],perr[2],er(-aeff**2 *kd/D/ka,popt[2])))
#print('Preb',1./(1+1./100),1-popt[0]/ka)
#print(popt[0]/kon)
```

```python
aeff = ka * (2 * rm) / (ka + kd)
def fitst2(tt,A,B,C):
    intk=A*(tt+B*sqrt(tt)+C)
    return [np.exp(-i*N/V) for i in intk]

fig,ax1=plt.subplots(1,1,figsize=(12,7))
lg=1
popt,pcov = curve_fit(fitst2,tt[lg:],st[lg:],p0=[kon,2*Bck,0],sigma=std[lg:])
plt.errorbar(tt,st,std,color='g')
#plt.plot(tt,st,'.')
plt.loglog(tt[lg:],fitst2(tt[lg:],*popt),'b--',label='fitted')
#plt.plot(tt,stt(intkt1),'r-',label='exact')
plt.legend()
#ax1.set_ylim(0.126,0.134)
er=lambda x,y:(x-y)*100/x
perr = np.sqrt(np.diag(pcov))
r=st-fitst2(tt,*popt)
meansq=sum(r**2)
chisq=sum((r/std)**2)
plt.ylim(0.95,1)
plt.xlim(0,0.005)
print(popt)
print(meansq,chisq)
print('kon={:.5},konfit={:.5}pm{:.5},diff={:.5},{:.5}%'.format(kon,popt[0],perr[0],kon-popt[0],er(kon,popt[0])))
print('Btheory={:.5},Bfit={:.5}pm{:.5},diff={:.5},{:.5}%'.format(2*Bck,popt[1],perr[1],2*Bck-popt[1],er(2*Bck,popt[1])))
#print('Ctheory={:.5},Cfit={:.5}pm{:.5},diff={:.5}%'.format(-aeff**2 *kd/D/ka,popt[2],perr[2],er(-aeff**2 *kd/D/ka,popt[2])))
#print('Preb',1./(1+1./100),1-popt[0]/ka)
#print(popt[0]/kon)
```

```python
def conc(fac,Na,Nb):
    rm=0.005
    L=fac*rm
    V=L**3
    Nt=Na+Nb
    v=4*3.14*rm**3/3
    print(Na*100*v/V,Nb*100*v/V,Nt*100*v/V)
```
