the zero order production rate kon must be divided by actual volume in Spatiocyte case in order to agree with ODE result.

```python
from ecell4 import *
from ecell4 import util
from ecell4.extra import ensemble
import math
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
rng = GSLRandomNumberGenerator()

duration = 1000
L=2
V = L**3
rm = 0.05
D=0.
wv=spatiocyte.SpatiocyteWorld.calculate_volume(Real3(L,L,L),rm)
kon= 1.
koff=0.5

def create_model(D,kon,rm): #sturock model
    with species_attributes():
        A | {'D': str(D), 'radius':str(rm)}
    with reaction_rules():
        ~A > A | kon
        #A > ~A | koff
    return get_model()

m1 = create_model(D,kon/V,rm)
m2 = create_model(D,kon/wv,rm)
#util.run_simulation(t=duration,y0={'A': 0,'B':N},volume=V, model=m, solver=("spatiocyte",rm),species_list=["A"])
obs1=ensemble.ensemble_simulations(duration,n=50, model=m2, solver=("spatiocyte",rm),volume=V,species_list=["A"],return_type='observer')#,return_type='observer')#,nproc=3,method='multiprocessing')#,return_type="observer")
obs2=util.run_simulation(t=duration,volume=V, model=m1, solver=("ode"),species_list=["A"],return_type='observer')
d1=np.array(obs1.data()).T
d2=np.array(obs2.data()).T
plt.plot(d1[0],d1[1],'.',label='spa')
plt.plot(d2[0],d2[1],'-',label='ode')
plt.title("zero_production of A")
plt.ylabel('number of A')
plt.xlabel('time')
plt.legend()
plt.savefig('/home/wx/wrk/ecellwork/figure/zero_order_production_spa_ODE.png')
```
