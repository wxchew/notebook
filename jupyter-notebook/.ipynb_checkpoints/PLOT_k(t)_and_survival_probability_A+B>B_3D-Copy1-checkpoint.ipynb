loading survival probability from spatiocyte simulation data of A+B>B, and calculate the k(t)

```python
#%matplotlib inline
from ecell4 import *
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import special
from scipy.optimize import curve_fit
from math import sqrt

rm = 0.005  
D = 1
kd= 4.0*math.pi*D*2*rm
r0=0.256318
f=1.0208582
C = 0.413497

def Phi(xx):
    return special.erfcx(xx)
    #return 1./( (377.*xx/324)+(1+ (314.*(xx**2)/847) )**0.5)

def getktt(ka,kdd,rm,tc):
    tau = ((ka*2*rm/(ka+kdd))**2)/(D)
    kt1 = np.array([(ka*kdd/(ka+kdd))*(1+(ka/kdd)*Phi(ka*(t/tau)**0.5 /kdd)) for t in tc])
    return kt1

def latkt(t,kon,B):
    return kon * (1 + B / np.sqrt(t))

def intkt(t,A,B):
    return A * (t + B * np.sqrt(t))


def calk(stsim,t,C):
    dst = (stsim[2:]-stsim[:-2])/(t[2:]-t[:-2])
    dst=np.append((stsim[1]-stsim[0])/(t[1]-t[0]),dst)
    dst=np.append(dst,(stsim[-1]-stsim[-2])/(t[-1]-t[-2])  )
    return dst/stsim/-C


error = lambda x,y: (x-y)*100/x

def plotk(dat,k_factor,voxelr,lag,rng,m,ts,Na,Nb,L):    
    if m=='o': mark='ori'
    else: mark = 'corrected'        
    tm =(2*voxelr)**2/(6*D)
    #print('tm',tm)
    w = spatiocyte.SpatiocyteWorld(ones()*L,voxelr)
    print('kfactor='+str(k_factor))
    ka = k_factor*kd
    #print('ka',ka)
    kon = (ka*kd)/(ka+kd)
    Vr = w.calculate_volume(ones()*L,voxelr)#43.46455449419419
    #print('Vactual',Vr,'V',L**3)        
    Pacc= ka/3/2**0.5/D/2/voxelr
    #print('Pacc',Pacc)
    kdd = 3*(2**0.5)*D*2*voxelr*(1/r0 -1)
    tau = ((ka*2*rm/(ka+kd))**2)/(D)
    aeff = ka * (2 * rm) / (ka + kd)
    Bck = aeff/math.sqrt(D*math.pi)
    Bt= 2*Bck       
    Cb=Nb/Vr
    std=np.loadtxt('/home/chew/outputdata/kfacc'+str(k_factor)+'_700_4000_4000_std_lag'+str(lag),delimiter=',')
    def intkt2(tt):
        return [kon*(t+(kd*tau/ka)*(2*ka*sqrt(t/tau/math.pi)/kd-1+Phi(ka*sqrt(t/tau)/kd))) for t in tt]
    
    def stt(intkt):
        return np.array([np.exp(-i*Cb) for i in intkt])
    
    def fitst2(t,A,B,C):
        return np.exp((A * (t + B * np.sqrt(t))+C )*(-Cb))
    
    tc=dat[:,0][::lag]
    mnac=dat[:,1]
    stsim = mnac[::lag]/Na

    #intkt = [kon*(t+(kd*tau/ka)*(2*ka*sqrt(t/tau/math.pi)/kd-1+Phi(ka*sqrt(t/tau)/kd))) for t in tc]
    #st = np.array([math.exp(-ik*Nb/Vr) for ik in intkt])
    #ksim = -(np.log(stsim[1:])-np.log(stsim[:-1]))/Cb/(tc[1:]-tc[:-1]) #k(t)=dlnS(t)/dt/cb 
    ksim = calk(stsim,tc,Cb)
    #ksim = -((stsim[1:])-(stsim[:-1]))/stsim[:-1]/Cb/(tc[1:]-tc[:-1]) #k(t)=dlnS(t)/dt/cb 
    tk = tc#[:-1]
    inx=(tk>ts[0])*(tk<ts[1])

    stck = stt(intkt(tc,kon,2*Bck))
    #stck = stt(intkt2(tc))
    #rt = stsim/stck
    #diff = stsim-stck
    #ks2=-(stsim[1:]-stsim[:-1])/stsim[:-1]/(tc[1:]-tc[:-1])/Cb
    ktck=latkt(tk,kon,Bck)
    #ktck=getktt(ka,kd,rm,tk)
    rkt = ksim/ktck
    #dkt = abs(ksim-ktck)
    #print('N(t)',mnac[:3])
    #print('stsim',stsim[:3])
    #print('ktsim',ksim[:3])
    #print('simt',tc[:3])

    NK=kon
    lw=3
    fs=22
    ms=10

    def ffit(ksim):
        popt,pcov = curve_fit(latkt,tk[inx],ksim[inx],p0=[kon,Bck],sigma=std[inx],absolute_sigma=True)#,bounds=(0,[np.inf,1,np.inf]))#,sigma=weight,absolute_sigma=False)
        perr = np.sqrt(np.diag(pcov))
        #print('fit result',popt)#,popt[0]/(3*2**0.5))
        r1=ksim[inx]-latkt(tk[inx],*popt)
        print('sq er fit1',sum(r1**2))
        print('kon={:.6},konfitted={:.6},error={:.6}%,pm{:.6},diff={:.6}'.format(kon,popt[0],error(kon,popt[0]),perr[0],kon-popt[0])) 
        aeffit = popt[1]*sqrt(math.pi*D)
        stdaeffit = perr[1]*sqrt(math.pi*D)
        print('B={:.6},Bfitted={:.6},error={:.6}%,pm{:.6},diff={:.6}'.format(Bck,popt[1],error(Bck,popt[1]),perr[1],Bck-popt[1]))  
        print('aeff={:.6},aeffitted={:.6},error={:.6}%,pm{:.6},diff={:.6}'.format(aeff,aeffit,error(aeff,aeffit),stdaeffit,aeff-aeffit))  
        #ax1.plot(tc[:-1],latkt(tc[:-1],*popt)/NK,'--',label='fitted',linewidth=lw)        
        return popt
    
    def fitst(st):
        popt,pcov = curve_fit(fitst2,tk[inx],st[inx],p0=[kon,2*Bck,0])#,bounds=(0,[np.inf,1,np.inf]))#,sigma=weight,absolute_sigma=False)
        perr = np.sqrt(np.diag(pcov))
        print('fit result',popt)#,popt[0]/(3*2**0.5))
        r1=st[inx]-fitst2(tk[inx],*popt)
        print('sq er fit1',sum(r1**2))
        print('kon={:.6},konfitted={:.6},error={:.6}%'.format(kon,popt[0],error(kon,popt[0]))) 
        print('B={:.6},Bfitted={:.6},error={:.6}%'.format(2*Bck,popt[1],error(2*Bck,popt[1])))  
        ax1.plot(tk[inx],fitst2(tk[inx],*popt),'--',label='fitted',linewidth=lw)        
        return popt    

    fit ='on'
    para='kt'
    if fit=='on':
        if para=='kt':
            popt1=ffit(ksim)
            #popt2=ffit(ksim2)
        elif para=='st':
            popt1=fitst(stsim)    

    if para=='kt':
        #kt
        #F=intk(tc,kon,Bt)
        #ax1.loglog(tk,rkt,label=str(k_factor)+'_'+str(tk[0]),alpha=0.5)
        #ax1.plot(tk,len(tk)*[1],'-')
        ax1.plot(tk,ksim/NK,'.',label='Simulation ($k_a/k_D=$'+str(k_factor)+')',markersize=ms)
        #ax1.errorbar(tk,ksim/NK,std/NK,label='L700, Sim $k_a/k_D=$'+str(k_factor),markersize=5)
        #ax1.loglog(tk,len(tk)*[kon/ka],'-')
        ax1.semilogx(tk,ktck/NK,'-',label='Collins-Kimball theory ($k_a/k_D=$'+str(k_factor)+')',markersize=5,linewidth=lw)
        #ax1.plot(tc[: -1], (F[1: ] - F[: -1]) / dtc / NK, 'c-',label='asymptotic CK theory',linewidth=lw)
        #ax1.plot(tc[:-1],latkt(tc[:-1],kon,Bck)/NK,'c-',label='CK theory',linewidth=lw)   
        #ax1.plot(tc[:-1],latkt(tc[:-1],*popt2)/NK,'--',label='fitted2',linewidth=lw)
        #ax1.plot(tt,ktck/NK,'c-',label='k(t) CK',linewidth=lw)
        ax1.set_ylim(0.98,1.2)
        ax1.set_ylabel('k(t)/keff',fontsize=fs)
        ax1.set_xlabel('time (s)',fontsize=fs)
          
    
    elif para=='st':    
        #st
        ax1.plot(tc,stck,'-',label='Collins-Kimball theory ($k_a/k_D=$'+str(k_factor)+')',linewidth=lw)
        ax1.plot(tc[inx],stsim[inx],'.',label='Simulation ($k_a/k_D=$'+str(k_factor)+')',markersize=ms)        
        ax1.set_ylabel('Survival probability',fontsize=fs)
        ax1.set_xlabel('time (s)',fontsize=fs)
        #ax1.semilogx(tc,rt,label='kfac'+str(k_factor))
  
    
    ax1.legend(loc='best',fontsize=fs-2)
    ax1.set_xlim(rng[0],rng[1])   

    #yrange=ks[(tk>rng[0]) *( tk<rng[1])]/NK
    #maxy=max(ksim[tk>rng[0]])/NK#yrange[0]
    #miny=min(ksim[tk<rng[1]])/NK#yrange[-1]
    #maxy2=max(lnS[tc>rng[0]])#yrange[0]
    #miny2=min(lnS[tc<rng[1]])#yrange[-1]    
    #print(maxy2,miny2)
    #ax1.set_ylim(miny,maxy)
    #ax1.set_ylim(1.5e-4,8e-4)
    #ax1.set_ylim(0.95,1.2)
    #ax1.set_ylim(0.01,0.0105)
    ax1.tick_params(labelsize=fs)
    print(' ')


    #plt.savefig("/home/chew/ecellwork/figure/ratioofst_sim_expected.png",format='png',dpi=300)            
```
